```
from bs4 import BeautifulSoup as bs
from nltk.corpus import stopwords,wordnet
from nltk.tokenize import sent_tokenize, word_tokenize
import string
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import requests
import nltk
import spacy
```

```
speech=requests.get("https://speakola.com/political/dr-a-p-j-abdul-kalam-vision-for-india-2011")
```

```
speech
```

```
soup=bs(speech.text,'html.parser')
```

```
soup
```

```
text=[]
for i in soup.find_all('p'):
  #print(i.text)
  text.append(i.text)
```

```
text
```

```
text=text[1:7]
```

```
text
```

```
fspeech=" ".join(str(x) for x in text)
fspeech
```

```
fspeech1=fspeech.translate(str.maketrans(' ',' ',string.punctuation))
fspeech1
```

```
nlp=spacy.load('en')
```

```
p_data=nlp(fspeech)
p_data
```

```
verb=[]
adj=[]
```

```
for i in p_data:
  #print(i.pos_)
  if i.pos_ == 'VERB':
    verb.append(i)
verb
```

```
for i in p_data:
  if i.pos_ == 'ADJ':
    adj.append(i)
adj
```
